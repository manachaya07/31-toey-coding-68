```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from datetime import datetime

# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏´‡∏±‡∏ß‡∏Ç‡πâ‡∏≠ (Title)
title = widgets.HTML(
    value="<h2 style='color: #FF69B4; text-align: center;'>üå∏ ‡πÇ‡∏õ‡∏£‡πÅ‡∏Å‡∏£‡∏°‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏≠‡∏≤‡∏¢‡∏∏‡∏™‡∏∏‡∏î‡∏ô‡πà‡∏≤‡∏£‡∏±‡∏Å üå∏</h2>",
)

# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ä‡πà‡∏≠‡∏á‡∏Å‡∏£‡∏≠‡∏Å‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• (Inputs)
birth_year_input = widgets.IntText(
    value=2000,
    description='‡∏õ‡∏µ‡πÄ‡∏Å‡∏¥‡∏î (‡∏Ñ.‡∏®.):',
    style={'description_width': 'initial'}
)

current_year_input = widgets.IntText(
    value=datetime.now().year,
    description='‡∏õ‡∏µ‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô:',
    style={'description_width': 'initial'}
)

# 3. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏õ‡∏∏‡πà‡∏°‡∏Å‡∏î‡∏™‡∏µ‡∏ä‡∏°‡∏û‡∏π (Pink Button)
btn_calculate = widgets.Button(
    description='‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏≠‡∏≤‡∏¢‡∏∏‡∏à‡πâ‡∏≤ ‚ú®',
    button_style='', # ‡∏õ‡∏¥‡∏î‡∏™‡πÑ‡∏ï‡∏•‡πå‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏™‡πà‡∏™‡∏µ‡πÄ‡∏≠‡∏á
    layout=widgets.Layout(width='200px', height='40px')
)
btn_calculate.style.button_color = '#FFB6C1' # ‡∏™‡∏µ‡∏ä‡∏°‡∏û‡∏π Light Pink

# 4. ‡∏™‡πà‡∏ß‡∏ô‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏• (Output)
output = widgets.Output()

def on_click_calculate(b):
    with output:
        clear_output()
        age = current_year_input.value - birth_year_input.value
        if age >= 0:
            print(f"‚ú® ‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡∏Ñ‡∏∏‡∏ì‡∏≠‡∏≤‡∏¢‡∏∏ {age} ‡∏õ‡∏µ‡πÅ‡∏•‡πâ‡∏ß‡∏ô‡∏∞‡∏à‡πä‡∏∞ ‚ú®")
        else:
            print("‚ö†Ô∏è ‡∏õ‡∏µ‡πÄ‡∏Å‡∏¥‡∏î‡∏ï‡πâ‡∏≠‡∏á‡πÑ‡∏°‡πà‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤‡∏õ‡∏µ‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô‡∏ô‡∏∞!")

btn_calculate.on_click(on_click_calculate)

# 5. ‡∏à‡∏±‡∏î‡∏ß‡∏≤‡∏á Layout ‡πÅ‡∏•‡∏∞‡πÅ‡∏™‡∏î‡∏á‡∏ú‡∏•
ui = widgets.VBox([title, birth_year_input, current_year_input, btn_calculate, output],
                  layout=widgets.Layout(align_items='center', border='2px solid #FFC0CB', padding='20px'))
display(ui)
```
